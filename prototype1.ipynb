```python
import pandas as pd

def cat_2_num(data_cat):
    # Iterate through each categorical column
    for col in data_cat.columns:
        # Perform one-hot encoding and rename columns
        one_hot = pd.get_dummies(data_cat[col], prefix=col, drop_first=True)
        
        # Drop the original categorical column from the DataFrame
        data_cat = data_cat.drop(col, axis=1)
        
        # Concatenate the one-hot encoded columns to the DataFrame
        data_cat = pd.concat([data_cat, one_hot], axis=1)
    
    return data_cat

# Example usage:
data_cat_encoded = cat_2_num(data_cat)
```
