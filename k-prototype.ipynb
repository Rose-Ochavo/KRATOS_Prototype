```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from kmodes.kprototypes import KPrototypes
```

```python
marketing_df = pd.read_csv('dataset2.csv')
```

```python
# Drop a single column
marketing_df = marketing_df.drop('Customer Id', axis='columns')

marketing_df['Defaulted'].fillna(0, inplace=True)  # Fill missing values with 0
```

```python
# marketing_df 
```

```python
# get the count of null values in each column
# marketing_df.isnull().sum()
```

```python
# marketing_df.describe()
```

```python
mark_array = marketing_df.values 
```

```python
# mark_array
```

```python
kproto = KPrototypes(n_clusters=3, verbose=2, max_iter=20)
clusters = kproto.fit_predict(mark_array, categorical = [7])
```

```python
# print(kproto.cluster_centroids_)
```

```python
cluster_dict = []
for c in clusters:
    cluster_dict.append(c)
```

```python
# cluster_dict
```

```python
marketing_df['cluster'] = cluster_dict
```

```python
# marketing_df
```

```python
# marketing_df[marketing_df['cluster']== 0].head(10)
```

```python
plt.scatter(marketing_df['Age'], marketing_df['Card Debt'], c=marketing_df['cluster'], cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Card Debt')
plt.title('Clustering Visualization')
plt.show()
```
