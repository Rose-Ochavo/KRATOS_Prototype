```python
import pandas as pd 

data = pd.read_csv('al.csv')
print(data.isnull().sum())

data = data.drop('ID', axis='columns')
data['QUANTITY'].fillna(1, inplace=True)  # Fill missing values with 0
data
```

```python
data.info()
```

```python
cols = ['TRANSACTION', 'BRAND', 'ITEM', 'PRICE', 'DATE']

# Use pd.get_dummies to create dummy variables for the specified columns
dummies = pd.get_dummies(data[cols])

# Concatenate the dummies with the original DataFrame
data = pd.concat([data, dummies], axis='columns')

# Drop the original columns since they are no longer needed
data.drop(columns=cols, inplace=True)

print(data)
```
