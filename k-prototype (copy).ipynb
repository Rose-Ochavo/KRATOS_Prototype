```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Step 1: Load the data from the CSV file
data = pd.read_csv('dataset.csv')

data['QUANTITY'].fillna(1, inplace=True)  # Fill missing values with 0

# Remove the ₱ symbol from the "PRICE" column
data['TRANSACTION'] = data['TRANSACTION'].str.replace('T', '')

# Remove the ₱ symbol from the "PRICE" column
data['PRICE'] = data['PRICE'].str.replace('₱', '')

# Step 2: Drop the 'ID' column
data = data.drop('ID', axis=1)

# Step 3: One-hot encode categorical variables
data = pd.get_dummies(data, columns=['BRAND', 'ITEM', 'DATE'])

# Step 4: Standardize the numeric variable (QUANTITY)
scaler = StandardScaler()
data['QUANTITY'] = scaler.fit_transform(data[['QUANTITY']])

# Step 5: Find the optimal number of clusters (k) using the Elbow Method
inertia = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Method to choose the optimal k
plt.figure(figsize=(8, 4))
plt.plot(range(2, 11), inertia, marker='o', linestyle='-', color='b')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.grid()
plt.show()

# Based on the Elbow Method, select an appropriate value for k

# Step 6: Perform k-means clustering with the selected k
k = 4  # You can choose the optimal k based on the Elbow Method
kmeans = KMeans(n_clusters=k, random_state=42)
data['Cluster'] = kmeans.fit_predict(data)

# Step 7: Calculate the Silhouette Score
silhouette_avg = silhouette_score(data, data['Cluster'])
print(f'Silhouette Score: {silhouette_avg}')

# Step 8: Create a scatter plot to visualize the clusters
plt.figure(figsize=(10, 6))
plt.scatter(data['ITEM'], data['PRICE'], c=data['Cluster'], cmap='viridis')
plt.xlabel('ITEM')
plt.ylabel('PRICE')
plt.title(f'K-Means Clustering (k={k}) - Silhouette Score: {silhouette_avg:.2f}')
plt.colorbar(label='Cluster')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.metrics import silhouette_score
from sklearn.decomposition import TruncatedSVD

# Load the CSV file
data = pd.read_csv('dataset.csv')

# Drop the 'ID' column
data = data.drop(columns=['ID'])

# Handle missing values in the 'QUANTITY' column
data['QUANTITY'].fillna(data['QUANTITY'].mean(), inplace=True)

# Separate numeric and categorical columns
numeric_cols = ['QUANTITY']
categorical_cols = ['TRANSACTION', 'BRAND', 'ITEM', 'PRICE', 'DATE']

# Create transformers for preprocessing
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])

# Combine preprocessing with K-means clustering
kmeans_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('reduce_dim', TruncatedSVD(n_components=2)),  # Using TruncatedSVD for sparse input
    ('cluster', KMeans())
])

# Determine the optimal number of clusters using Silhouette Score
silhouette_scores = []
k_values = range(2, 11)
for k in k_values:
    kmeans_pipeline.set_params(cluster__n_clusters=k)
    kmeans_pipeline.fit(data)
    labels = kmeans_pipeline.predict(data)
    silhouette_scores.append(silhouette_score(data, labels))

# Plot Silhouette Score vs. Number of Clusters
plt.plot(k_values, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs. Number of Clusters')
plt.grid(True)
plt.show()

# Based on the plot, choose the optimal number of clusters

# Fit K-means with the selected number of clusters
optimal_num_clusters = 3  # You can change this based on the plot
kmeans_pipeline.set_params(cluster__n_clusters=optimal_num_clusters)
kmeans_pipeline.fit(data)

# Get cluster labels
labels = kmeans_pipeline.predict(data)

# Add cluster labels to the original dataset
data['Cluster'] = labels

# Scatter plot of the clusters
plt.scatter(data['QUANTITY'], data['PRICE'], c=data['Cluster'], cmap='rainbow')
plt.xlabel('QUANTITY')
plt.ylabel('PRICE')
plt.title('K-means Clustering')
plt.show()
```
